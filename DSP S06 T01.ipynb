### Mostra sistemàtica

```python
# em creo un dataframe buit per despres anar incorporant els registres
sport_msis=pd.DataFrame(columns=sport.columns)
```

```python
t=int(input('Tamany de la mostra:'))
k=len(sport)/t
print(f"K={k}")
```

```python
base=(random.uniform(1,k))
#print(0,base,round(base))
for i in range(t):
    indice=base+(i*k)
    print(i+1,indice,round(indice))
    #print(sport.iloc[round(indice)])
    sport_msis=sport_msis.append(sport.iloc[round(indice)])
```

## Exercici 3

Continua amb el conjunt de dades de tema esportiu i genera una mostra utilitzant el mètode Reservoir sampling.

```python
sport_reservoir=pd.DataFrame(columns=sport.columns)
```

```python
def generator(max):
    number = 1
    while number < max:
        number += 1
        yield number
# Doing Reservoir Sampling
k=10
reservoir = []
for i, element in enumerate(sport.index):
    if i+1<= k:
        reservoir.append(element)
    else:
        probability = k/(i+1)
        if random.random() < probability:
            # Select item in stream and remove one of the k items already selected
             reservoir[random.choice(range(0,k))] = element
print(reservoir)
```

### Mostra Reservoir Sampling

```python
for i in reservoir:
    sport_reservoir=sport_reservoir.append(sport.iloc[i])
sport_reservoir
```
